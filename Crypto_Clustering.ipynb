```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Get the columns' names
df_market_data.columns
```

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data_scaled = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 
                                                                       'price_change_percentage_7d',
                                                                       'price_change_percentage_14d', 
                                                                       'price_change_percentage_30d',
                                                                       'price_change_percentage_60d', 
                                                                       'price_change_percentage_200d',
                                                                       'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
df_market_data_transformed = pd.DataFrame(df_market_data_scaled, columns=['price_change_percentage_24h', 
                                                                       'price_change_percentage_7d',
                                                                       'price_change_percentage_14d', 
                                                                       'price_change_percentage_30d',
                                                                       'price_change_percentage_60d', 
                                                                       'price_change_percentage_200d',
                                                                       'price_change_percentage_1y'])
# Copy the crypto names from the original data
df_market_data = df_market_data.reset_index()
df_coin_id = df_market_data['coin_id']
df_market_data_transformed= pd.concat([df_coin_id ,df_market_data_transformed], axis = 1)

# Set the coinid column as index
df_market_data_transformed = df_market_data_transformed.set_index('coin_id')

# Display sample data
df_market_data_transformed.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
print(k)
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k

for i in k:
# Inside the loop:

# 1. Create a KMeans model using the loop counter fo r the n_clusters
    model = KMeans(n_clusters = i, random_state=1)
# 2. Fit the model to the data using `df_market_data_transformed`
    model.fit(df_market_data_transformed)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)

print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_dict)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_plt = elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)

elbow_plt
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4 clusters, as more clasters than 4 doesn't give us much better fit to the data

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = model.predict(df_market_data_transformed)

# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame
df_market_data_transformed_predicted = df_market_data_transformed.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_transformed_predicted['crypto_clusters'] = crypto_clusters

# Display sample data
df_market_data_transformed_predicted.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

cluster_plt = df_market_data_transformed_predicted.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    color="crypto_clusters", 
    hover_cols=["coin_id"],   
)

cluster_plt
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca_data = pca.fit_transform(df_market_data_transformed_predicted)

# View the first five rows of the DataFrame. 
crypto_pca_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
variance = pca.explained_variance_ratio_
print(variance)
```

```python
# Calculate the sum of variances
total_variance = sum(variance)
total_variance
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89% is the total explained variance of the three principal components. 

```python
# Create a new DataFrame with the PCA data.
df_crypto_pca =  pd.DataFrame(crypto_pca_data, columns=["PC1", "PC2", "PC3"])


# Copy the crypto names from the original data
df_crypto_pca = pd.concat([df_coin_id, df_crypto_pca], axis = 1)

# Set the coinid column as index
df_crypto_pca = df_crypto_pca.set_index('coin_id')

# Display sample data
df_crypto_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
     model = KMeans(n_clusters = i, random_state=1)
# 2. Fit the model to the data using `df_crypto_pca`
     model.fit(df_crypto_pca)
# 3. Append the model.inertia_ to the inertia list
     inertia.append(model.inertia_)

print(inertia)

```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca_dict = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_pca_dict)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_pca_plt = elbow_pca_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)

elbow_pca_plt
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it doesn't differ from the k-value found using the original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(df_crypto_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_pca_clusters = model_pca.predict(df_crypto_pca)

# Print the resulting array of cluster values.
print(crypto_pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_crypto_pca_predicted = df_crypto_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_crypto_pca_predicted['crypto_pca_clusters'] = crypto_pca_clusters

# Display sample data
df_crypto_pca_predicted.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.


# Visualization 1: PC1 vs PC2
cluster_pca_plt = df_crypto_pca_predicted.hvplot.scatter(
    x="PC1",
    y="PC2",
    color="crypto_pca_clusters", 
    hover_cols=["coin_id"],   
    xlabel="PC1",   
    ylabel="PC2",  
)

cluster_pca_plt
```
